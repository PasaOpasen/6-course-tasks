# Import packages

```python
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import StandardScaler


from keras.models import Sequential
from keras.layers import Dense
```

# Data reading

```python
data = pd.read_csv('../input/introduction-to-deep-learning-nn-with-keras/concrete_data.csv')

data
```

# Fitting models

```python
scores = []

matrix = data.to_numpy()

for _ in range(50):
    
    X_train, X_test, y_train, y_test = train_test_split(matrix[:,:-1], matrix[:, -1], test_size = 0.3)
    
    sc = StandardScaler()
    sc.fit(X_train)
    X_train = sc.transform(X_train)
    X_test = sc.transform(X_test)
    
    model = Sequential()
    
    model.add(Dense(10, input_dim=X_train.shape[1], activation='relu'))
    model.add(Dense(1))
    
    model.compile(loss='mean_squared_error', optimizer='adam')
    
    model.fit(X_train, y_train, epochs = 50)
    
    scores.append(mean_squared_error(y_test, model.predict(X_test)))
```

# See results

```python
scores = np.array(scores)

print(f'mean of errors: {np.mean(scores)}; std of errors: {np.std(scores)}')
```
